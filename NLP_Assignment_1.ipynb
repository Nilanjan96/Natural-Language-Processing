The function to extract the text. Note that I removed the .lower() function which changed all the characters to lower case. It was because since we would work to understand the semantic meaning of the sentence, capital letters in certain positions are important

````
 def extract_text(filename,output_as_list=False):
    '''
    Extract the text from the file name (json file) and
    index the content from paper_id, title, abstract and body_text fields
    Retuns - if output_as_list = True
                list of all the above values,
             else
                text of title, abstract and bodt_text
    '''

    file = open('/content/drive/MyDrive/pdf_json/'+filename)
    body_text = ""
    abstract = ""
    title = ""
    paper_id = ""

    paper_content = json.load(file)

    #get the paper_id
    if 'paper_id' in paper_content:
        paper_id = paper_content['paper_id']
    #get the title, if available
    if 'title' in paper_content['metadata']:
        title = paper_content['metadata']['title']
    #get abstract.text, if available
    if 'abstract' in paper_content:
        for abs in paper_content['abstract']:
            abstract = abstract + abs['text']
    if 'body_text' in paper_content:
        for bt in paper_content['body_text']:
            body_text = body_text + bt['text']


    if output_as_list:
            return [paper_id,title,abstract,body_text]
    else:
        return (title + ' ' + abstract + ' ' + body_text + ' ')

````

For pre processing, we use 3 steps

Step 1) Clean the data from non ascii characters-
        Since we are focusing on ascii characters and english papers to build a 
        corpus and understand its semantic, we remove all non ascii characters
        from the dataset

Step 2) Cleaning the data from html-
        htmls and urls to websites have no use in our understanding of the     corpus.Thus we use regex to find and clean the dataset off html links.

Step 3) Cleaning the data from numbers-
        Since we are focusing on understanding the semantic meaning of the corpus, numbers are of no use to us. Note that we do not remove the numbers which are attached to words like COVID-19. We only remove numbers and decimals which are not connected to any word through a regex filter.

Defining the function to remove any non ascii character from a string

````
def non_ascii_clear(string): 
    answer = ''.join(filter(lambda x: (ord(x) <= 128), string))
    return answer.strip()
````

````
#Example:-
non_ascii_clear(' ')
````

Defining the function to remove any html from a string

````
def html_clear(html_given): 
    regex = re.compile('<.*?>|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});')
    cleaned_text = re.sub(regex, '', html_given)
    return cleaned_text.strip()
````

````
#Example:- 
html_clear('<a href="http://www.dr-chuck.com/page2.htm">')
````

Defining the function to remove any number from a string

````
def number_clear(number): 
    regex = re.compile('^\d*[.,]?\d*$')
    cleaned_text = re.sub(regex, '', number)
    return cleaned_text.strip()
````

````
#Example:-
number_clear('134.532')
````

We combine the pre processing steps mentioned above to make this function which basically takes a list of line (containing title, abstract and body text) and gives back a single line delimited by '#' for every single document in the corpus. We then add this line to the corpus.

````
def clean_text(line):
  #first we clean the title, then the abstract and lastly the body text.
  final = ''
  title = html_clear(line[0])
  title = non_ascii_clear(title)
  title = number_clear(title)
  #if(len(title) != 0): 
   # sc += 1
    #wc += len(title.split(' '))
  #now we clear abstract
  abstract = html_clear(line[1])
  abstract = non_ascii_clear(abstract)
  abstract = number_clear(abstract)
  #if(len(abstract) != 0): 
   # split_abstract = abstract.split(".")
    #sc += len(split_abstract)
    #for i in split_abstract:
     # wc += len(i.strip().split(' '))

  #now we clear body text
  body = html_clear(line[2])  
  body = non_ascii_clear(body)
  body = number_clear(body)
  #if(len(body) != 0): 
   # split_body = body.split(".")
    #sc += len(split_body)
    #for i in split_body:
     # wc += len(i.strip().split(' '))

  final = title.strip() + '# ' + abstract.strip() + '# ' + body.strip() + '\n'
  return final
````

**Question 5**

To count the number of words in the corpus, we need to remove all the numbers and special characters from the corpus. Here we have defined this function so as to just take a line, and just return the alphabets with spaces in that line, which will help us to count the words

````
def clearnonalpha(line): 
    whitelist = set('abcdefghijklmnopqrstuvwxyz ABCDEFGHIJKLMNOPQRSTUVWXYZ')
    line = ''.join(filter(whitelist.__contains__, line))
    return line
````

Next we read the corpus. Every document is delimited by a newline so we make an array to store all the documents. We will now count the number of words in every document and add them to find the word_count of the total document. Similarly we will do the same for the number of sentences.
We will also keep an array number_of_words_cum_sum which is a cumulative sum of the number of words per document. This will help us plot the graphs later.

````
corpus = open('corpus.txt','r')
corpus1 = corpus.read()
corpus.close()
````

The last line is empty, so we pop it out

````
documents = corpus1.split('\n')
documents.pop()
````

````
word_count = 0
sentence_count = 0
number_documents= 0
number_of_words_cum_sum = []
for i in documents:
  print(number_documents)
  number_documents += 1
  num_words = clearnonalpha(i).split() #ignores puncuations 
  word_count = word_count + len(num_words)
  number_of_words_cum_sum.append(word_count)
  threeparts = i.split('#')
  for j in threeparts:
    num_sentences = j.split(".") #ignores puncuations 
    sentence_count = sentence_count + len(num_sentences)
````

**Question 6**

To count the vocabulary, which is basically number of unique words, we use the same idea as above, we just use a set this time. We also use an array vocabulary_cum_sum which is the cumulative sum of the number of words in the vocabulary of each document

````
vocab = set()
vocabulary_cum_sum = []
number_documents = 0
number_vocab = 0
for i in documents:
  print(number_documents)
  number_documents += 1
  num_words = clearnonalpha(i).split() #ignores puncuations 
  for i in num_words:
    if(i not in vocab): number_vocab += 1
    vocab.add(i.lower())
  vocabulary_cum_sum.append(number_vocab)
````

Now, we plot log Terms vs log Vocabulary graph

````
plt.figure(figsize = (15,10))
plt.plot(np.log(number_of_words_cum_sum),np.log(vocabulary_cum_sum))
plt.xlabel("Number of words")
plt.ylabel("Vocabulary")
plt.show()
````

Next we plot the log estimated vocabulary by heaps law vs log actual vocabulary graph

````
heap_law_list = []
for i in number_of_words_cum_sum:
  heap_law_list.append(30*(i)**0.6)
````

````
plt.figure(figsize = (15,10))
plt.plot(np.log(number_of_words_cum_sum),np.log(vocabulary_cum_sum),label = 'actual vocabulary')
plt.plot(np.log(number_of_words_cum_sum),np.log(heap_law_list),label = 'estimated vocabulary through heap\'s law')
plt.xlabel("Number of words")
plt.ylabel("Vocabulary")
plt.legend()
plt.show()
````
